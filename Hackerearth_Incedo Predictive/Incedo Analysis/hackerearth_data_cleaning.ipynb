```
train['score_3'].fillna(train['score_3'].mean(), inplace=True)
train['score_4'].fillna(train['score_4'].mean(), inplace=True)
train['score_5'].fillna(train['score_5'].mean(), inplace=True)
#test['score_3'].fillna(test['score_3'].mean(), inplace=True)
```

```
train['Essayset'].fillna(method='ffill',inplace=True)
```

```
train['clarity'].unique()
```

```

train['clarity'].loc[train['clarity']=='worst'] = 0
train['clarity'].loc[train['clarity']=='average'] = 1
train['clarity'].loc[train['clarity']=='above_average'] = 2
train['clarity'].loc[train['clarity']=='excellent'] = 3
```

```
train['clarity'].fillna(train['clarity'].mean(), inplace=True)
```

```
train['coherent'].unique()
```

```
train['coherent'].loc[train['coherent']=='worst'] = 0
train['coherent'].loc[train['coherent']=='average'] = 1
train['coherent'].loc[train['coherent']=='above_average'] = 2
train['coherent'].loc[train['coherent']=='excellent'] = 3
```

```
train['coherent'].fillna(train['coherent'].mean(), inplace=True)
```

```
test['clarity'].loc[test['clarity']=='worst'] = 0
test['clarity'].loc[test['clarity']=='average'] = 1
test['clarity'].loc[test['clarity']=='above_average'] = 2
test['clarity'].loc[test['clarity']=='excellent'] = 3
test['coherent'].loc[test['coherent']=='worst'] = 0
test['coherent'].loc[test['coherent']=='average'] = 1
test['coherent'].loc[test['coherent']=='above_average'] = 2
test['coherent'].loc[test['coherent']=='excellent'] = 3
test['clarity'].fillna(test['clarity'].mean(), inplace=True)
test['coherent'].fillna(test['coherent'].mean(), inplace=True)
```
