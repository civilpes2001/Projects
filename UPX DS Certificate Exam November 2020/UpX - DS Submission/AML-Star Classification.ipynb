## Splitting training dataset into train and test

```
X = df.copy().drop(['TargetClass'], axis = 1)
y = df['TargetClass']
```

```
# Splitting training dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

**Scaling the data**

```
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc_fit = sc.fit(X_train)
X_train = sc_fit.transform(X_train)
X_test = sc_fit.transform(X_test)
```

# 1. XGBoost Model

```
#XGBoost Model
from xgboost import XGBClassifier
from sklearn import metrics

xgb = XGBClassifier()
xgb_fit = xgb.fit(X_train, y_train)
```

**Predictions**

```
y_xgb_pred_test = xgb_fit.predict(X_test)
y_xgb_pred_train = xgb_fit.predict(X_train)
```

**Train and Test accuracy**

```
print(metrics.accuracy_score(y_test, y_xgb_pred_test))#test accuracy
print(metrics.accuracy_score(y_train, y_xgb_pred_train))#train accuracy
```

**Confustion Matrix**

```
metrics.confusion_matrix(y_test, y_xgb_pred_test)
```

**Classification report**

```
cl = classification_report(y_test,y_xgb_pred_test)
print(cl)
```
