# Train Test Split

```
X = df.drop('prognosis',axis=1).values
Y = df['prognosis'].values
```

```
#np.unique([1, 1, 2, 2, 3, 3])
values = np.unique(Y)
values
```

```
y = np.zeros(shape = (len(Y),len(values)))
k=0
for x in Y:
     for i in range(len(values)):
            if x==values[i]:
                tmp=list(np.zeros(41))
                tmp[i]=1
                y[k]=tmp
                k+=1
```

```
y[0]
```

```
np.unique(y)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
```

```
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```
model = Sequential()
model.add(Dense(units = 3000, input_dim=132, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 1500, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 41, activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer='adam')
```

```
model.fit(x=X_train, 
          y=y_train, 
          epochs=3,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```
pre = model.predict_proba(X_test)
```

```
pre[0].max()
```

```
acc=0;
for i in range(len(pre)):
    if pre[i].argmax() == y_test[i].argmax() :
        if pre[i].max() >= 0.90:
            acc+=1

acc=acc/len(pre)
acc
```
